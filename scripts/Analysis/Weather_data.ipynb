```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('C:\\Users\\rajpu\\OneDrive\\Documents\\Projects_git\\forest-fire-prediction\\extracted\\FInal_weather\\cariboo_master_weather_2012-01-01_2023-12-31.csv')
```

```python
for column in df.columns:
    plt.figure(figsize=(10, 6))
    plt.hist(df[column].dropna(), bins=30, edgecolor='k')
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.grid(axis='y', alpha=0.75)
    plt.show()
```
