```python
import requests
import pandas as pd
import io
import urllib.parse
```

```python
WFS_URL = "https://openmaps.gov.bc.ca/geo/pub/wfs"

LAYER_HISTORICAL = "pub:WHSE_LAND_AND_NATURAL_RESOURCE.PROT_HISTORICAL_FIRE_POLYS_SP"
LAYER_CURRENT = "pub:WHSE_LAND_AND_NATURAL_RESOURCE.PROT_CURRENT_FIRE_POLYS_SP"


REGION_NAME = "Cariboo Fire Centre"
START_YEAR = 2001
CURRENT_YEAR = 2024
```

```python
def get_wfs_data(layer_name, cql_filter=None):
    """
    Fetches data from the BC WFS service and returns a Pandas DataFrame.
    """
    params = {
        "service": "WFS",
        "version": "2.0.0",
        "request": "GetFeature",
        "typeNames": layer_name,
        "outputFormat": "csv",
        "srsName": "EPSG:4326",  # Latitude/Longitude
    }
    
    if cql_filter:
        params["CQL_FILTER"] = cql_filter

    print(f"Requesting data from layer: {layer_name}...")
    try:
        response = requests.get(WFS_URL, params=params, timeout=60)
        response.raise_for_status()
        
        # Check if empty
        if not response.content:
            print("No data returned.")
            return pd.DataFrame()
            
        # Parse CSV content
        df = pd.read_csv(io.StringIO(response.text))
        return df
    except Exception as e:
        print(f"Error fetching data: {e}")
        return pd.DataFrame()
```

```python
cql_hist = f"FIRE_YEAR >= {START_YEAR}"
df_hist = get_wfs_data(LAYER_HISTORICAL, cql_hist)
```

```python
if not df_hist.empty:
        df_hist.columns = [c.upper() for c in df_hist.columns]
if not df_curr.empty:
        df_curr.columns = [c.upper() for c in df_curr.columns]
```

```python
all_fires = pd.concat([df_hist, df_curr], ignore_index=True)
```

```python
all_fires.dtypes
```

```python
print(f"Total records fetched  {len(all_fires)}")
```

```python
fc_col = next((col for col in ['FIRE_CENTRE', 'MOF_FIRE_CENTRE_NAME'] if col in all_fires.columns), None)
```

```python
fc_col
```

```python
if fc_col:
        print(f"Filtering by {fc_col} for 'Cariboo'...")
        cariboo_df = all_fires[all_fires[fc_col].astype(str).str.contains("Cariboo", case=False, na=False)]
else:
        print("Could not identify Fire Centre column. Saving all data.")
        cariboo_df = all_fires
```

```python
wanted_columns = [
        'FIRE_NUMBER', 'FIRE_YEAR', 'FIRE_CAUSE', 'SIZE_HA', 
        'LATITUDE', 'LONGITUDE', 'FIRE_DATE', 'IGNITION_DATE', fc_col
    ]
```

```python
final_cols = [c for c in wanted_columns if c in cariboo_df.columns]
final_df = cariboo_df[final_cols]
```

```python
filename = "BC_Cariboo_Wildfires_2001-2024.csv"
```

```python
final_df.to_csv(filename, index=False)
```
